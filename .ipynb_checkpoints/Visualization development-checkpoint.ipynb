```python
def gen_surface_for_matlab(surface, data_man):
    template_path = '/Users/murat/Dropbox/Projects/pylib/data/templates/templates_32k/surface/'
    surf_file = Gifti(template_path + surface + '_32k_L.surf.gii',
                      template_path + surface + '_32k_R.surf.gii')
    data2write = {'vertices_l': surf_file.data(0)[0],
                'vertices_r': surf_file.data(0)[1],
                'triangles_l': surf_file.data(1)[0],
                'triangles_r': surf_file.data(1)[1]}
    data_man.save('templates/templates_32k/surface/' + surface + '_32k.mat', data2write)
```

```python
gen_surface_for_matlab('Conte69_inflated', data_man)
gen_surface_for_matlab('Conte69_midthickness', data_man)
gen_surface_for_matlab('Conte69_veryinflated', data_man)
```
